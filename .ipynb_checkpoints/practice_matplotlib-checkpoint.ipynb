```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
#plt.show()
# ^ will show the plot - should not be used more then onece per session
```

```python
# Create the plot with ax.plt()  using the object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis) 
```

```python
# Create the plot with ax.plt() using the object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
#Switch the axis so that the Fare($) data is on the x-axis.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)

#Change the bar color to cyan.
ax.barh(x_axis, y_axis, color= “tab:cyan”)

#Add Title and axes labels
ax.set_xlable(“Date”)
ax.set_ylable(“Fare($)“)
ax.set_xlable(“PyBer Fare by Month”)

#Invert the y-axis
ax.gca().invert_yaxis()

#legend
fig.legend([“Chicago”])
```
